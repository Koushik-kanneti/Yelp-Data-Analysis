## PART 5: 10 SQL queries

### 5.1: Top 5 cities with most restaurants
#### This query helps us see which cities have the most restaurant-related businesses, giving insights into competitive density and potential markets.

```python
%%sql
SELECT TOP 5 city, COUNT(*) AS num_restaurants
FROM business
WHERE categories LIKE '%Restaurant%'
GROUP BY city
ORDER BY num_restaurants DESC;

```

### 5.2 Find Average Star Rating of Restaurants in the City
#### Provides insight into the general quality and customer satisfaction level of restaurants in each of the top 5 cities.

```python
%%sql
SELECT city, ROUND(AVG(stars),2) AS avg_rating
FROM business
WHERE city IN ('Tampa', 'Philadelphia', 'Indianapolis', 'Nashville', 'Tucson') 
  AND categories LIKE '%Restaurant%'
GROUP BY city
ORDER BY avg_rating DESC;
```

### 5.3: Average Number of Fans for Yelp Users in the City
#### Helps gauge the general engagement and influence of Yelp users in the top 5 city, which is valuable for planning marketing and making restaurant choice.

```python
%%sql
SELECT business.city, AVG(users.fans) AS avg_fans
FROM users
JOIN review ON users.user_id = review.user_id
JOIN business ON review.business_id = business.business_id
WHERE business.city IN ('Tampa', 'Philadelphia', 'Indianapolis', 'Nashville', 'Tucson')
GROUP BY business.city
ORDER BY avg_fans DESC;
```

### 5.4: Find Categories with the Most Reviews in a City
#### Reveals popular restaurant categories in a chosen city, giving insights into the most sought-after cuisines.

```python
%%sql
SELECT TOP 10 categories, SUM(review_count) AS total_reviews
FROM BUSINESS
WHERE city = 'Tampa'
GROUP BY categories
ORDER BY total_reviews DESC;
```

### 5.5: Identify Popular Restaurant Categories in the City
#### Helps identify popular restaurant categories in the city, allowing us to understand what types of restaurants are common and possibly find a niche.

```python
%%sql
SELECT TOP 5 categories, COUNT(*) AS num_businesses
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%'
GROUP BY categories
ORDER BY num_businesses DESC;

```

### 5.6: Top 10 High-Rated Restaurant Categories
#### Shows high-rated restaurant types, helping you decide which categories might attract quality-conscious customers.

```python
%%sql
SELECT TOP 10 categories, AVG(stars) AS avg_stars
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%'
GROUP BY categories
ORDER BY avg_stars DESC;
```

### 5.7: Commonly Reviewed Restaurants in the City
#### Identifies popular restaurants based on review count, which can signal high customer interest or strong marketing

```python
%%sql
SELECT TOP 10 name, review_count
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%'
ORDER BY review_count DESC;
```

### 5.8: Number of Open vs. Closed Restaurants
#### Shows the ratio of open to closed restaurants, potentially indicating market saturation or challenges.

```python
%%sql
SELECT 
    is_open,
    COUNT(*) AS num_restaurants,
    ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM business WHERE city = 'Tampa' AND categories LIKE '%Restaurant%')),2) AS percentage
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%'
GROUP BY is_open;

```

### 5.9: Identify Areas in Tampa with High-Rated Restaurants
#### Pinpoints postal codes with high-rated restaurants, helping in selecting an area known for quality dining options.

```python
%%sql
SELECT TOP 10 postal_code, ROUND(AVG(stars),2) AS avg_stars
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%'
GROUP BY postal_code
ORDER BY avg_stars DESC;
```

### 5.10: Average Rating by Restaurant Attributes
#### Shows which attributes (e.g., "Good for Groups," "Outdoor Seating") are associated with higher ratings, providing insight into customer preferences.

```python
%%sql
SELECT 
    ROUND(AVG(CASE WHEN attributes LIKE '%"BusinessAcceptsCreditCards": "True"%' THEN stars END),2) AS avg_stars_credit_cards,
    ROUND(AVG(CASE WHEN attributes LIKE '%"RestaurantsDelivery": "True"%' THEN stars END),2) AS avg_stars_delivery,
    ROUND(AVG(CASE WHEN attributes LIKE '%"OutdoorSeating": "True"%' THEN stars END),2) AS avg_stars_outdoor_seating,
    ROUND(AVG(CASE WHEN attributes LIKE '%"RestaurantsTakeOut": "True"%' THEN stars END),2) AS avg_stars_takeout,
    ROUND(AVG(CASE WHEN attributes LIKE '%"BikeParking": "True"%' THEN stars END),2) AS avg_stars_bike_parking
FROM business
WHERE city = 'Tampa' AND categories LIKE '%Restaurant%';
```
